```python
import matplotlib.pyplot as plt
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,rwidth=0.8)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,rwidth=0.8,bins=[80,100,125,150],color=['purple'])
plt.xlabel('sugar level')
plt.ylabel('no of patients')
plt.title('blood sugar level')
```

```python
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar,blood_sugar_women],bins=[80,100,125,150],color=['orange','red'],label=['men','women'])
plt.legend()
plt.xlabel('sugar level')
plt.ylabel('no of patients')
plt.title('blood sugar level')
```
