```python
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.4,autopct='%0.2f%%')
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.4,autopct='%0.2f%%',explode=[0,0.12,0,0,0])
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.4,autopct='%0.2f%%',startangle=90)
```
