```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[42,56,14,8,45,47,32]
```

```python
plt.plot(x,y)
```

```python
plt.xlabel('day')
plt.ylabel('temp')
plt.title('weather')
plt.plot(x,y,linewidth=5,color='red')
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,49,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel('day')
plt.ylabel('temp')
plt.title('weather')

plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')

plt.legend(loc='best')
```
